```python
X=data.drop(['Exited'],axis=1)
y=data.Exited
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,train_size=0.2)
```

## Requried function for implementation of logistic Regression

```python
# Sigmoid function
def sigmoid(z):
    return 1 / (1 + np.exp(-z))

# Forward propagation
def forward_propagation(X, weights, bias):
    # Linear combination
    z = np.dot(X, weights) + bias
    # Sigmoid activation
    y_predicted = sigmoid(z)
    return y_predicted

# Backward propagation
def backward_propagation(X, y, y_predicted):
    n_samples = X.shape[0]
    # Calculate gradients
    dw = (1 / n_samples) * np.dot(X.T, (y_predicted - y))
    db = (1 / n_samples) * np.sum(y_predicted - y)
    return dw, db

# Loss function (Binary cross-entropy)
def compute_loss(y, y_predicted):
    n_samples = y.shape[0]
    # Avoid division by zero by adding a small value epsilon
    epsilon = 1e-15
    y_predicted = np.clip(y_predicted, epsilon, 1 - epsilon)
    # Binary cross-entropy loss
    loss = -1 / n_samples * np.sum(y * np.log(y_predicted) + (1 - y) * np.log(1 - y_predicted))
    return loss

# Update weights and bias
def update_parameters(weights, bias, dw, db, learning_rate):
    weights -= learning_rate * dw
    bias -= learning_rate * db
    return weights, bias
```


## model building




```python

# Initialize weights and bias
weights = np.zeros(x_train.shape[1])
bias = 0

# Set hyperparameters
learning_rate = 0.1
num_iterations = 1000

# Training loop
for i in range(num_iterations):
    # Forward propagation
    y_predicted = forward_propagation(x_train, weights, bias)

    # Backward propagation
    dw, db = backward_propagation(x_train, y_train, y_predicted)

    # Update weights and bias
    weights, bias = update_parameters(weights, bias, dw, db, learning_rate)

    # Calculate and print loss every 100 iterations
    if i % 100 == 0:
        loss = compute_loss(y_train, y_predicted)
        print(f"Iteration {i}: Loss = {loss}")

# Predictions on test set
y_test_predicted = forward_propagation(x_test, weights, bias)
y_test_predicted_class = (y_test_predicted >= 0.5).astype(int)

# Evaluate model
accuracy = np.mean(y_test_predicted_class == y_test)
print(f"Accuracy: {accuracy}")
```
